### 1.2 Scatter Plot
A scatter plot is a set of points plotted on horizontal and vertical axes. The scatter plot can be used to study the correlation between the two variables. One can also detect the extreme data points using a scatter plot.


```python
# check the head() of the dataset
df.head()
```

Plot the scatter plot for the variables 'total_bill' and 'tip'.

```python
# data
X = df['1stmathstest']  # Assign the values from the '1stmathstest' column of the DataFrame df to X.
Y = df['2ndmathstest']  # Assign the values from the '2ndmathstest' column of the DataFrame df to Y.

# Create a scatter plot using X and Y
plt.scatter(X, Y)  

# add the axes labels to the plot
plt.xlabel('first test of maths')
plt.ylabel('second test of maths')

# Display the scatter plot
plt.show()  
```

We can add different colors, opacity, and shapes to data points. Let's add these customizations to the above plot.

```python
# plot the scatter plot for the variables 'total_bill' and 'tip'

X = df['1stmathstest']  # Assign the values from the '1stmathstest' column of the DataFrame df to X.
Y = df['2ndmathstest']  # Assign the values from the '2ndmathstest' column of the DataFrame df to Y.


# plot the scatter plot
# s is for shape, c is for colour, alpha is for opacity (0 < alpha < 1)
plt.scatter(X, Y, s = np.array(Y)**1.9, c= 'green', alpha= 0.5)

# add title 
plt.title('Scatter Plot')

# add the axes labels to the plot
plt.xlabel('first test of maths')
plt.ylabel('second test of maths')

# display the plot
plt.show()
```

The bubbles with greater radius display that the tip amount is more as compared to the bubbles with less radius.

                    <b>Boxplot is a way to visualize the five-number summary of the variable. The five-number summary includes the numerical quantities like minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum. Boxplot gives information about the outliers in the data. Plot the boxplot of 'total_bill' to check the distribution and presence of outliers in the variable.

```python
# plot a distribution of total bill
plt.boxplot(df['1stmathstest'])

# add labels for five-number summary
plt.text(x=1.1, y=df['1stmathstest'].min(), s='min')
plt.text(x=1.1, y=df['1stmathstest'].quantile(0.25), s='Q1')
plt.text(x=1.1, y=df['1stmathstest'].median(), s='median (Q2)')
plt.text(x=1.1, y=df['1stmathstest'].quantile(0.75), s='Q3')
plt.text(x=1.1, y=df['1stmathstest'].max(), s='max')

# add the graph title and axes labels
plt.title('Boxplot of 1st test of maths grades')
plt.ylabel('1st test maths grades')

# display the plot
plt.show()
```

The above boxplot clearly shows the presence of outliers above the horizontal line. We can add an arrow to showcase the outliers. Also, the median (Q2) is represented by the orange line, which is near to Q1 rather than Q3. This shows that the distribution of the grade is simetric.

Lets plot a distribution plot of '3rdmathstest'.

```python
# plot a distribution plot
sns.displot(df['3rdmathstest'], kind='kde')

# display the plot
plt.show()
```

```python
# # iterate the distplot() function over the sex

# # list of sex
sex = ['male', 'female']

# # iterate through sex
for i in sex:
    subset = df[df['sex'] == i]
    
# # Draw the density plot`
# # 'hist = False' will not plot a histogram
# # 'kde = True' plots density curve
    sns.distplot(subset['3rdmathstest'], hist = False, kde = True,
                  kde_kws = {'shade':True},
                  label = i)

# Set plot labels and add a legend
plt.xlabel('3rd Math Test Scores')
plt.ylabel('Density')
plt.legend(title='Sex')
```

```python
# # iterate the distplot() function over the sex

# # list of sex
sex = ['male', 'female']

# # iterate through sex
for i in sex:
    subset = df[df['sex'] == i]
    
# # Draw the density plot`
# # 'hist = False' will not plot a histogram
# # 'kde = True' plots density curve
    sns.distplot(subset['2ndmathstest'], hist = True, kde = False,
                  kde_kws = {'shade':True},
                  label = i)

# Set plot labels and add a legend
plt.xlabel('3rd Math Test Scores')
plt.ylabel('Density')
plt.legend(title='Sex')
```
