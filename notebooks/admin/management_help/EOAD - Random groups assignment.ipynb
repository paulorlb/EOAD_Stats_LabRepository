```python
df_alunos_eoad_group1 = df_alunos_eoad[df_alunos_eoad.SELECT_STATS_WORK_1 == 1]
df_alunos_eoad_group2 = df_alunos_eoad[df_alunos_eoad.SELECT_STATS_WORK_1 == 2]
```

```python
df_alunos_eoad_group1.iloc[[0,-1]]
```

```python
# Shuffle the student names column
df_alunos_eoad_group1_shuffled = df_alunos_eoad_group1.sample(frac=1, random_state=41).reset_index(drop=True)
```

```python
# Total number of students
n_students = df_alunos_eoad_group1_shuffled.shape[0]

# Determine number of groups of 3 and groups of 2
n_groups_of_3 = n_students // 3
remaining_groups_students = n_students % 3

print(n_groups_of_3, remaining_groups_students)

# If there are remaining students (1 or 2), adjust the number of groups
if remaining_groups_students == 1:
    # If there is 1 student left over, break one group of 3 into two groups of 2
    n_groups_of_3 -= 1
    n_groups_of_2 = 2
elif remaining_groups_students == 2:
    # If there are 2 students left over, they form one group of 2
    n_groups_of_2 = 1
else:
    n_groups_of_2 = 0

# Assign group numbers: start by assigning students to groups of 3
group_numbers = np.concatenate([
    np.repeat(np.arange(n_groups_of_3), 3),
    np.repeat(np.arange(n_groups_of_3, n_groups_of_3 + n_groups_of_2), 2)
])

# Add the group assignment to the DataFrame
df_alunos_eoad_group1_shuffled['Group'] = group_numbers
```

```python
df_alunos_eoad_group1_shuffled[['Nome', 'NMecAluno', 'Group']].to_csv('EOAD2526_Group1_Assignments.csv', index=False)
```
